```python
import os
import os.path
import cv2
import math
import numpy as np
```

```python
def clahe(image_path, denoise=False, limit=None):
    bgr = cv2.imread(image_path)
    lab = cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB)
    lab_planes = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=limit, tileGridSize=(16,32))
    lab_planes[0] = clahe.apply(lab_planes[0])
    lab = cv2.merge(lab_planes)
    bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
    return bgr
```

```python
def convert(image_dir, output_home, limit=None):
    i = 1
    list = os.listdir(image_dir)
    file_count = len(list)
    for f in list:
        if ("ppm" not in f) and ('jpg' not in f):
            continue
        if i >= file_count:
            break
        print(i)
        i += 1
        print("processing: %s" % (f))
        
        image_path = os.path.join(image_dir, f)
        bgr = clahe(image_path, limit=limit)
        basename = os.path.basename(image_dir)
        output_dir = os.path.join(output_home, basename)
        
        if not os.path.isdir(output_dir):
            os.mkdir(output_dir)
        output_file = os.path.join(output_dir, f)
        cv2.imwrite(output_file, bgr)
```

```python
# untuk dir, tulis nama foldernya. file ini harus 1 folder sama folder sourcenya
input_home_dir = "../img_src"
output_home_dir = "../img_res"
limit = 2.0
```

```python
list_dir = os.listdir(input_home_dir)
convert(input_home_dir, output_home_dir, limit=limit)
```

```python
def he(image_dir, output_home):
    i = 1
    list = os.listdir(image_dir)
    file_count = len(list)
    for f in list:
        if ("ppm" not in f) and ('jpg' not in f):
            continue
        if i >= file_count:
            break
        print(i)
        i += 1
        print("processing: %s" % (f))
        
        image_path = os.path.join(image_dir, f)   
        img = cv2.imread(image_path)

        img_y_cr_cb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
        y, cr, cb = cv2.split(img_y_cr_cb)

        img_eq = cv2.equalizeHist(y)

        img_y_cr_cb_eq = cv2.merge((img_eq, cr, cb))
        img_rgb_eq = cv2.cvtColor(img_y_cr_cb_eq, cv2.COLOR_YCR_CB2BGR)
        
        basename = os.path.basename(image_dir)
        output_dir = os.path.join(output_home, basename)
        
        if not os.path.isdir(output_dir):
            os.mkdir(output_dir)
        output_file = os.path.join(output_dir, f)
        cv2.imwrite(output_file, img_rgb_eq)
        
he("../img_src", "../img_he")
```

```python
def calculatePSNR(img1, img2):
    mse = np.mean( (img1 - img2) ** 2 )
    if mse == 0:
        return 100
    PIXEL_MAX = 255.0
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))
```

```python
#Calculate PSNR for STARE dataset
ori_dir = '../img_src'
clahe_dir = '../img_he/img_src' #change it to he or clahe
ori_list = os.listdir(ori_dir)
clahe_list = os.listdir(clahe_dir)
psnr_stare = []
i = 0
for f in ori_list:
    if ("ppm" in f):
        i += 1
        ori_path = os.path.join(ori_dir, ori_list[i])
        original = cv2.imread(ori_path)
        clahe_path = os.path.join(clahe_dir, clahe_list[i])
        clahe = cv2.imread(clahe_path)
        psnr_stare.append(calculatePSNR(original, clahe))  
        
print(sum(psnr_stare)/len(psnr_stare))
```
